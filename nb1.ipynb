```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor()
```

```python
train= pd.read_csv('train.csv')
train
```

```python
test=pd.read_csv('test.csv')
test
```

```python
X1_train=train.drop(['T2M','QV2M','VACATION_RATE'],axis=1)
X1_train
```

```python
y1_train=train["QV2M"]
y1_train
```

```python
#to predict QV2M using train
regressor.fit(X1_train,y1_train)
pred1=regressor.predict(X1_train) 
pred1.round(1)
```

```python
y1_train.round(1)
```

```python
accuracy=regressor.score(X1_train,y1_train)
accuracy
```
